```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import mean_squared_error, r2_score
# Cargar el conjunto de datos
data = pd.read_csv(r'C:\Users\Lenovo\OneDrive\Desktop\6 SEMESTRE\Analisis de datos\heart_cleveland_upload.csv')

#C:\Users\Lenovo\OneDrive\Desktop\6 SEMESTRE\Analisis de datos\heart_cleveland_upload.csv
```

```python
X = data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca']]  
```

```python
# Crear y usando la columna 'condition'

y = data['condition']
```

```python
#dividir los datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  
```

```python
# Definir las variables predictoras (X) y la variable objetivo (y)
X = data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope']]
y = data['condition']
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
print('\nReporte de Clasificación:')
print(classification_report(y_test, y_pred))
```

```python
print('\nMatriz de Confusión:')
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión')
plt.ylabel('Valores Reales')
plt.xlabel('Valores Predichos')
plt.show()
```
